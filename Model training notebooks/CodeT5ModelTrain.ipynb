```python
from datasets import Dataset,DatasetDict


train_dataset = Dataset.from_dict(dataset)
val_dataset = Dataset.from_dict(validation)
test_dataset = Dataset.from_dict(test)
my_dataset_dict = DatasetDict({"train":train_dataset,"test":val_dataset})
print(type(train_dataset))
```

```python
max_input_length = 500
max_target_length = 500
source_lang = "en"
target_lang = "java"

prefix = "Translate: "
def preprocess_examples(examples):
    pseudocodes = examples['pseudocode']
    cppcode =examples['code']
   
    inputs = [prefix + each for each in pseudocodes]
    model_inputs = tokenizer(inputs, max_length=max_input_length, padding="max_length", truncation=True)

    labels = tokenizer(cppcode, max_length=max_target_length, padding="max_length", truncation=True).input_ids

    labels_with_ignore_index = []
    for labels_example in labels:
        labels_example = [label if label != 0 else -100 for label in labels_example]
        labels_with_ignore_index.append(labels_example)

    model_inputs["labels"] = labels_with_ignore_index

    return model_inputs
```

```python


train_dataset = train_dataset.map(preprocess_examples, batched=True)
val_dataset = val_dataset.map(preprocess_examples, batched=True)
test_dataset = test_dataset.map(preprocess_examples, batched=True)
```

```python
sweep_config = {
    'method': 'random'
    }
metric = {
    'name': 'loss',
    'goal': 'minimize'   
    }

sweep_config['metric'] = metric
parameters_dict = {
 'warmup_steps':{
     'values': [500,1000,1500]
 },
    
    }

sweep_config['parameters'] = parameters_dict
parameters_dict.update({
    'epochs': {
        'value': 1}
    })

parameters_dict.update({
    'learning_rate': {
        # a flat distribution between 0 and 0.1
        'distribution': 'uniform',
        'min': 5e-5,
        'max': 1e-3
      }
      
    })
import pprint

pprint.pprint(sweep_config)
```

```python
from transformers import T5ForConditionalGeneration, AdamW, get_linear_schedule_with_warmup
import pytorch_lightning as pl

class CodeT5(pl.LightningModule):
    # def __init__(self, lr=5e-5, num_train_epochs=1, warmup_steps=1000,max_epochs=1):
    def __init__(self, lr=5e-5, num_train_epochs=1, warmup_steps=1000,max_epochs=1):
        super().__init__()
        self.model = T5ForConditionalGeneration.from_pretrained("Salesforce/codet5-small")
        self.save_hyperparameters()

    def forward(self, input_ids, attention_mask, labels=None):     
        outputs = self.model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
        return outputs
    
    def common_step(self, batch, batch_idx):
        outputs = self(**batch)
        loss = outputs.loss

        return loss
      
    def training_step(self, batch, batch_idx):
        loss = self.common_step(batch, batch_idx)     
        # logs metrics for each training_step,
        # and the average across the epoch
        self.log("training_loss", loss)
        # self.log('train_accuracy', self.accuracy(logits, y))
        return loss

    def validation_step(self, batch, batch_idx):
        loss = self.common_step(batch, batch_idx)     
        self.log("validation_loss", loss, on_epoch=True)

        return loss

    def test_step(self, batch, batch_idx):
        loss = self.common_step(batch, batch_idx)     

        self.log("test_loss", loss, on_epoch=True)

        return loss

    def configure_optimizers(self):
        # create optimizer
        optimizer = AdamW(self.parameters(), lr=self.hparams.lr)
        # create learning rate scheduler
        num_train_optimization_steps = self.hparams.num_train_epochs * len(train_dataloader)
        lr_scheduler = {'scheduler': get_linear_schedule_with_warmup(optimizer,
                                                    num_warmup_steps=self.hparams.warmup_steps,
                                                    num_training_steps=num_train_optimization_steps),
                        'name': 'learning_rate',
                        'interval':'step',
                        'frequency': 1}
        
        return {"optimizer": optimizer, "lr_scheduler": lr_scheduler}

    def train_dataloader(self):
        return train_dataloader

    def val_dataloader(self):
        return valid_dataloader

    def test_dataloader(self):
        return test_dataloader
```

```python
from pytorch_lightning import Trainer

from pytorch_lightning.loggers import WandbLogger
from pytorch_lightning.callbacks import EarlyStopping, LearningRateMonitor

def train():
    
    
    with wandb.init(config=None):
        # If called by wandb.agent, as below,
        # this config will be set by Sweep Controller
        
        config = wandb.config
        model = CodeT5(lr=config.learning_rate,warmup_steps = config.warmup_steps)
        wandb_logger = WandbLogger(name='codet5-finetune-translate', project='CodeT5')
# for early stopping, see https://pytorch-lightning.readthedocs.io/en/1.0.0/early_stopping.html?highlight=early%20stopping
        early_stop_callback = EarlyStopping(
            monitor='validation_loss',
            patience=3,
            strict=False,
            verbose=False,
            mode='min'
        )


        lr_monitor = LearningRateMonitor(logging_interval='step')

        trainer = Trainer(gpus=1, 
                          max_epochs=1,
                        #   default_root_dir="/content/drive/MyDrive/CodeT5/Notebooks/Checkpoints", 
                          logger=wandb_logger, 
                          callbacks=[early_stop_callback, lr_monitor])
        trainer.fit(model)
wandb.agent(sweep_id, train, count=5)
```
